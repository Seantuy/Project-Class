**Comparing the complexity of different algorithms**

O(1) - Constant time

```python
def get_element(arr, index):
    return arr[index]
```

O(logn) - Binary search

```python
def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

*O(n) - Linear complexity*

Looking for the maximum value of an array

```python
def find_max(arr):
    if len(arr) == 0:
        raise ValueError("Input array is empty")

    # Initialize the maximum value with the first element
    max_value = arr[0]

    # Iterate through the array, updating the maximum value
    for element in arr:
        if element > max_value:
            max_value = element

    return max_value
```

Timeit, as a mean to observe time complexity experimentally

```python
import timeit
import random

# Définition de la fonction find_max
def find_max(arr):
    max_value = arr[0]
    for element in arr:
        if element > max_value:
            max_value = element
    return max_value

# Générer des vecteurs de différentes tailles pour les tests
sizes = [10, 100, 1000, 10000]
vectors = {size: [random.randint(1, 10000) for _ in range(size)] for size in sizes}

# Mesurer le temps d'exécution pour chaque taille de vecteur
for size in sizes:
    vector = vectors[size]
    time_taken = timeit.timeit(lambda: find_max(vector), number=1000)  # Mesurer le temps d'exécution sur 1000 exécutions
    print(f"Taille du vecteur : {size}, Temps d'exécution moyen : {time_taken / 1000:.6f} secondes")
```

O(n log n) - linearithmic complexity


The example of Merge sort.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

O(n^2) - Quadratic complexity

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

*Constant time complexity*

Function: get_element(arr, index)
Objective: This function retrieves an element from an array (or list) based on its index.
How it works: The function simply returns the array element at the specified index. Since accessing the array by index is a constant-time operation in Python, the time complexity is O(1).

*O(log n) - Logarithmic time complexity*

Function: binary_search(arr, x)
Purpose: This function searches for a value x in a sorted array (or list) and returns its index if found, otherwise -1.
How it works: The function divides the search interval in half on each iteration. If the middle element is the target value, it returns the index. If the target value is less than the middle element, the search continues in the left half; otherwise, it continues in the right half. This division process continues until the value is found or the interval is empty.

*O(n) - Linear time complexity*

Function: find_max(arr)
Purpose: This function finds and returns the maximum value in an array (or list).
How it works: The function initializes the maximum value with the first element of the array. It then iterates through the array, updating the maximum value each time it encounters a larger element.

*O(n log n) - Linearithmic time complexity*

Function: merge_sort(arr)
Purpose: This function sorts an array (or list) using the merge sort algorithm.
How it works: The function recursively splits the array into two halves until each subarray contains a single element. Then it merges these subarrays in sorted order. The merge function is used to combine two sorted arrays into a single sorted array.

*O(n^2) - Quadratic time complexity*

Function: bubble_sort(arr)
Purpose: This function sorts an array (or list) using the bubble sort algorithm.
How it works: The function repeatedly iterates through the list, compares adjacent elements, and swaps them if they are not in the correct order. This process is repeated for each element in the list, ensuring that the largest unsorted element "moves up" to its correct position with each iteration.
Each of these algorithms serves a specific purpose, and their time complexities give us an idea of ​​how their performance scales with the size of the input data.

*Worst vs best scenario*

Bubble Sort:

Worst case (O(n^2)):
Scenario: The input array is in reverse order.
Example data: [5, 4, 3, 2, 1]
Explanation: In this case, each element must be compared and exchanged on the first pass, then all but one on the second pass, and so on. This results in the maximum number of comparisons and swaps, leading to a quadratic number of operations.

Average case (O(n^2)):
Scenario: The input array is in random order.
Example data: [3, 5, 1, 4, 2]
Explanation: On average, bubble sort will always perform a quadratic number of comparisons and swaps. Although this may not always be as numerous as in the worst case, the order of growth remains quadratic.

Best case (O(n)):
Scenario: The input array is already sorted.
Example data: [1, 2, 3, 4, 5]
Explanation: In the best case, bubble sort will perform n-1 comparisons on the first pass and realize that no swaps are needed, indicating that the list is already sorted. Although some implementations of bubble sort continue to make unnecessary passes, an optimized version may detect that the list is sorted and terminate prematurely, resulting in a linear number of operations.

To truly illustrate the performance differences for these cases, one could measure the actual execution time of the algorithm on these datasets using Python's timeit module or by counting the number of compares and swaps. However, the above scenarios give a conceptual understanding of how bubble sort performance varies with different sets of input data

```python
# Optimal scenarios 

import random

# Worst-case: Reverse order
worst_case = list(range(1000, 0, -1))

# Average-case: Random order
average_case = list(range(1, 1001))
random.shuffle(average_case)

# Best-case: Already sorted
best_case = list(range(1, 1001))
```

```python
import timeit

# Measure time for worst-case
worst_case_time = timeit.timeit(lambda: bubble_sort(worst_case.copy()), number=10)
print(f"Worst-case time: {worst_case_time:.5f} seconds")

# Measure time for average-case
average_case_time = timeit.timeit(lambda: bubble_sort(average_case.copy()), number=10)
print(f"Average-case time: {average_case_time:.5f} seconds")

# Measure time for best-case
best_case_time = timeit.timeit(lambda: bubble_sort(best_case.copy()), number=10)
print(f"Best-case time: {best_case_time:.5f} seconds")
```

**Exponential complexity: the Fibonacci example**

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

The diagram below will show you why the Fibonacci sequence falls under this time complexity

```python
fibonacci(5)
|-- fibonacci(4)
|   |-- fibonacci(3)
|   |   |-- fibonacci(2)
|   |   |   |-- fibonacci(1)
|   |   |   |-- fibonacci(0)
|   |   |-- fibonacci(1)
|   |-- fibonacci(2)
|       |-- fibonacci(1)
|       |-- fibonacci(0)
|-- fibonacci(3)
    |-- fibonacci(2)
    |   |-- fibonacci(1)
    |   |-- fibonacci(0)
    |-- fibonacci(1)
```

            = 1 + 0
            = 1


```python
fibonacci(4) = fibonacci(3) + fibonacci(2)
            = (fibonacci(2) + fibonacci(1)) + fibonacci(2)
            = (1 + 1) + 1
            = 3
```

```python
fibonacci(5) = fibonacci(4) + fibonacci(3)
             = (fibonacci(3) + fibonacci(2)) + (fibonacci(2) + fibonacci(1))
             = ((fibonacci(2) + fibonacci(1)) + fibonacci(2)) + (fibonacci(2) + fibonacci(1))
             = ((1 + 1) + 1) + (1 + 1)
             = 5
```

```python
import timeit

# Define a function to measure the runtime of fibonacci for a given input size
def measure_time(n):
    # Use a lambda function to pass the argument to fibonacci
    time_taken = timeit.timeit(lambda: fibonacci(n), number=1)
    return time_taken

# Measure runtime for different input sizes
input_sizes = list(range(1, 31))  # Adjust the range as needed
times = [measure_time(n) for n in input_sizes]

# Print the results
for n, time_taken in zip(input_sizes, times):
    print(f"fibonacci({n}) took {time_taken:.6f} seconds")
```

**Let us shift to spatial complexity**

Linear Space Complexity - O(n):

Algorithm: duplicate an array.

```python
def duplicate_array(arr):
    copy = []
    for num in arr:
        copy.append(num)
    return copy
```

Binary search : Logarithmic complexity - O(log n):

```python
def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

The algorithm divides the search interval in half at each iteration. Although the time complexity is O(log n), the space complexity is O(1) because it uses a constant amount of space (the variables low, high, and mid). However, if this were implemented recursively, the space complexity would be O(log n) due to the depth of the call stack.

Exponential complexity - O(2^n):

```python
def power_set(s):
    if not s:
        return [[]]
    subsets = power_set(s[:-1])
    next_elem = s[-1]
    return subsets + [item + [next_elem] for item in subsets]
```

For a set of size n, there are 2^n possible subsets. The algorithm recursively calculates these subsets, resulting in exponential growth in the space used.

When analyzing space complexity, it is essential to consider both the space used by the variables and data structures of the algorithm and the space used by the call stack (for recursive algorithms).
